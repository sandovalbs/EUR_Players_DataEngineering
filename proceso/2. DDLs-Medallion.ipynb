```sql
create widget text storageLocation default "abfss://unit-catalog@datlkeempresa.dfs.core.windows.net";
create widget text catalogo default "catalog_empresa";
```

```sql
use catalog ${catalogo}
```

### DDLs Tablas

```sql
CREATE OR REPLACE TABLE ${catalogo}.bronze.CLIENTES (
  id INT, nombre STRING, correo STRING, pais STRING
)
USING DELTA
LOCATION '${storageLocation}/bronze/CLIENTES';

CREATE OR REPLACE TABLE ${catalogo}.bronze.VENTAS (
  id INT, id_cliente INT, fecha DATE, monto DOUBLE
)
USING DELTA
LOCATION '${storageLocation}/bronze/VENTAS';

CREATE OR REPLACE TABLE ${catalogo}.bronze.PRODUCTOS (
  id INT, nombre STRING, categoria STRING, precio DOUBLE
)
USING DELTA
LOCATION '${storageLocation}/bronze/PRODUCTOS';
```

### Creacion tablas Silver

```sql
CREATE OR REPLACE TABLE ${catalogo}.silver.VENTAS_LIMPIAS (
  id INT, id_cliente INT, fecha DATE, monto DOUBLE
)
USING DELTA
LOCATION '${storageLocation}/silver/VENTAS_LIMPIAS';

CREATE OR REPLACE TABLE ${catalogo}.silver.CLIENTES_VALIDOS (
  id INT, nombre STRING, correo STRING, pais STRING, activo BOOLEAN
)
USING DELTA
LOCATION '${storageLocation}/silver/CLIENTES_VALIDOS';

CREATE OR REPLACE TABLE ${catalogo}.silver.DETALLE_PRODUCTOS (
  id INT, nombre STRING, categoria STRING, precio DOUBLE, stock INT
)
USING DELTA
LOCATION '${storageLocation}/silver/DETALLE_PRODUCTOS';
```
